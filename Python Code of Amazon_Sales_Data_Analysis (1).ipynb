```python
data.groupby(['Region','Sales Channel'])['Total Profit'].sum()
```

```python
year_wise_sales = data.groupby('Year')['Total Revenue'].sum()
plt.figure(figsize=(9,5))
sns.barplot(x=year_wise_sales.index , y=year_wise_sales.values)
plt.title('Year Wise Sales avg:')
plt.xlabel('Year')
plt.ylabel('Total Revenue')
plt.show()
```

```python
Total_revenue_itemtype = data.groupby('Item Type')['Total Revenue'].sum()
```

```python
plt.figure(figsize=(9,5))
Total_revenue_itemtype.plot(kind ='bar')
plt.title('Avg Revenue by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Revenue')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(9,5))
region_total_revenue = data.groupby('Region')['Total Profit'].mean()
plt.pie(region_total_revenue, startangle=90 ,labels = region_total_revenue.index , autopct='%1.1f%%' )
plt.title('Average profit in Region wise')
plt.show()
#
```

```python
year_month_units_sold = data.groupby(['Year','Month'])['Units Sold'].sum()
```

```python
plt.figure(figsize=(7,7))
year_month_units_sold.plot(kind ='bar')
plt.title('Units Sold by Year and Month')
plt.xlabel('Year and Month')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.grid(axis='y')
```

```python
Total_revenue_sales_channel = data.groupby('Sales Channel')['Total Revenue'].mean()
```

```python
plt.figure(figsize=(8,8))
plt.tight_layout()
Total_revenue_sales_channel.plot(kind='pie',autopct = '%1.1f%%',startangle = 90)
plt.title('Total Revenue Sales Channel')
plt.show()
#
```

```python
Region_units_sold = data.groupby('Region')['Units Sold'].sum()
plt.figure(figsize=(7,7))
Region_units_sold.plot(kind='pie',autopct = '%1.1f%%',startangle = 90)


#

center_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.title('Units sold by Region')
plt.axis('equal')
#
```

```python
Total_cost_sales_channel = data.groupby('Sales Channel')['Total Cost'].sum()
```
